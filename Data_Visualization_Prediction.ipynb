# **MÔ HÌNH Linearregression**

```python
#Xem thông tin các biến trong dữ liệu
df.info()
```

## **Bước 1: xác định biến đầu ra (output), các biến đầu vào (inputs)**

```python
#Mô hình 1: có 2 biến đầu vào là 'M2', 'INF' và 1 biến đầu ra là 'GDP'
y1=df['GDP']
X1=df[['M2', 'INF']]

#mô hình 2: có 4 biến đầu vào là 'M2', 'INF', 'R', 'TRADE' và 1 biến đầu ra là 'GDP'
y2=df['GDP']
X2=df[['M2', 'INF', 'R', 'TRADE']]
```

## **Bước 2: chia tập dữ liệu**

```python
#Gọi hàm train_test_split để chia tập dữ liệu
from sklearn.model_selection import train_test_split

#Chia X1,y1 của mô hình 1
X1_train,X1_test,y1_train,y1_test=train_test_split(X1,y1,test_size=0.1,random_state=15)

#Chia X2,y2 của mô hình 2
X2_train,X2_test,y2_train,y2_test=train_test_split(X2,y2,test_size=0.1,random_state=15)
```

## **Bước 3: khởi tạo mô hình, dùng thuật nào để huấn luyện mô hình**

```python
#Gọi mô hình LinearRegression
from sklearn.linear_model import LinearRegression

#Mô hình 1
model1=LinearRegression()

#Mô hình 2
model2=LinearRegression()
```

## **Bước 4: huấn luyện mô  hình với training data: X_train, y_train**

```python
#Huấn luyện mô hình 1
model1.fit(X1_train,y1_train)

#Huấn luyện mô hình 2
model2.fit(X2_train,y2_train)
```

```python
model1.coef_
```

```python
model1.intercept_
```

## **Bước 5: kiểm định mức độ dự báo chính xác của mô hình**

```python
#Dự báo cho mô hình 1 trên tập test
y_dubao1=model1.predict(X1_test)

#Dự báo cho mô hình 2 trên tập test
y_dubao2=model2.predict(X2_test)

#Tạo bảng kết quả
import pandas as pd
predict_y=pd.DataFrame({'y_test':y1_test,'y_dubao1':y_dubao1,'y_dubao2':y_dubao2})
predict_y
```

```python
#Nhập các thư viện
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error

#tính các đại lượng MSE, RMSE, MAE cho mô hình 1
mse1=mean_squared_error(y1_test,y_dubao1)
rmse1=np.sqrt(mse1)
mae1=mean_absolute_error(y1_test,y_dubao1)

#tính các đại lượng MSE, RMSE, MAE cho mô hình 2
mse2=mean_squared_error(y2_test,y_dubao2)
rmse2=np.sqrt(mse2)
mae2=mean_absolute_error(y2_test,y_dubao2)

# Tạo DataFrame cho các chỉ số đánh giá
accuracy_df = pd.DataFrame({
    'Metric': ['MSE', 'RMSE', 'MAE'],
    'Model 1': [mse1, rmse1, mae1],
    'Model 2': [mse2, rmse2, mae2]
})

print('====Mức độ dự báo chính xác của 2 mô hình====')
accuracy_df
```

# **MÔ HÌNH VAR**

```python
#Tải dữ liệu các chỉ số từ World Bank data để tiến hành phân tích: GDP per capita growth (annual %), Broad money (% of GDP); Inflation, consumer prices (annual %); Real interest rate (%); Trade (% of GDP)

from pandas_datareader import wb

df=wb.download(indicator=['NY.GDP.MKTP.KD.ZG', 'FM.LBL.BMNY.GD.ZS','FP.CPI.TOTL.ZG','FR.INR.RINR','NE.TRD.GNFS.ZS'], country=['VNM'], start=2010, end=2024)

df.columns=['GDP', 'M2', 'INF','R','TRADE']

df
```

```python
#df ban đầu có thêm cột quốc gia, chỉ nên lấy cột year để vẽ
df.index=df.index.get_level_values('year')
df=df.fillna(df.mean())
df
```

```python
df.index = pd.date_range(start='2010', periods=len(df), freq='YS')
df.index=df.index.year
df
```

```python
# Nhập thư viện
from statsmodels.tsa.api import VAR

# Chia train/test
n = int(len(df) * 0.9)
train = df[:n]
test = df[n:]

# Fit VAR
model = VAR(train)
results = model.fit(maxlags=1, ic='aic')

# Dự báo
lag = results.k_ar
forecast_extended = results.forecast(train.values[-lag:], steps=len(test) + 2)

# Tạo index: từ năm đầu test đến hết 2 năm thêm
start_year = test.index[0]
forecast_ext_df = pd.DataFrame(forecast_extended, columns=df.columns,
                                index=range(start_year, start_year + len(test) + 2))

forecast_ext_df
```
